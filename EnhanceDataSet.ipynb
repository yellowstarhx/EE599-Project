```
# Creat enhanced training set

from PIL import Image
import matplotlib.pyplot as plt
import os
from skimage import io,data

path = '/content/gdrive/Team Drives/EE599 Project/Phase 1/enhanced_train/'
for i in range(1, 41):
  class_path = path + str('{:03}'.format(i)) + '/'
  for j in range(1, 6):
    img_path = class_path + str(j) + '.png'
    if(os.path.isfile(img_path)==0):
      img_path = class_path + str(j) + '.jpg'
    img = Image.open(img_path)
    # Transpose
    transposed  = img.transpose(Image.FLIP_TOP_BOTTOM)
    io.imsave(class_path + str(j * 5 + 1) + '.png', transposed)
    transposed  = img.transpose(Image.FLIP_LEFT_RIGHT)
    io.imsave(class_path + str(j * 5 + 2) + '.png', transposed)
    # Rotate it by 45 degrees
    rotated = img.rotate(45)
    io.imsave(class_path + str(j * 5 + 3) + '.png', rotated)
    rotated = img.transpose(Image.ROTATE_270)
    io.imsave(class_path + str(j * 5 + 4) + '.png', rotated)
    rotated = transposed.rotate(45)
    io.imsave(class_path + str(j * 5 + 5) + '.png', rotated)
      
```
